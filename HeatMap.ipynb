Price changes have powerful effects on both the company and the consumer. In order to analyze a pricing change I would ensure that we were tracking every potential impact so that we could make a decision informed across all dimensions.

1) Company revenue -  I would make sure revenue tracking is in place.

2) Signups and retention would be critical for measuring the long term impact of a pricing change - I would create retention heat maps - below. We could track the impact of price changes across cohorts. I would also suggest using a small hold out group for both new and existing customers to ensure we were measuring the impact of our price change and not impacts from other changes.

3) Test website interactions with the pricing page and monitor whether there are any observable impacts on lead creation for people who viewed the pricing page, whether bounce rates have have changed, and ensure that any impact on new user behavior was accounted for.

4) I would work in close collaboration with our Customer Success and Sales teams to make sure that any qualitative feedback we received was taken into account and that we understood the real world reaction to product changes.




```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

C_range = 10. ** np.arange(-2, 8)
gamma_range = 10. ** np.arange(-5, 4)

confMat=np.random.rand(10, 9)

heatmap = plt.pcolor(confMat)

for y in range(confMat.shape[0]):
    for x in range(confMat.shape[1]):
        plt.text(x + 0.5, y + 0.5, '%.2f' % confMat[y, x],
                horizontalalignment='center',
                verticalalignment='center',)


#plt.grid() #this will look bad now
plt.colorbar(heatmap)
plt.subplots_adjust(left=0.15, right=0.99, bottom=0.15, top=0.99)
plt.ylabel('Cost')
plt.xlabel('Gamma')

plt.xticks(np.arange(len(gamma_range))+0.5, gamma_range, rotation=45,)
plt.yticks(np.arange(len(C_range))+0.5, C_range, rotation=45)
plt.show()
```
