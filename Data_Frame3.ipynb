Review of the .apply() method it applys an operation to every value within that series

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS!"
bond["Box Office"].apply(convert_to_string_and_add_millions) ### changes each value in this series per the function above, which you pass in as an argument to the apply
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_millions)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions) # column to extracted on every iteration
```

how to combine values in a row in a custom function

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
# create a column that based off values in every single row -- CODE BELOW IS AWSOME !!!!!!
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have non clue"
    
bond.apply(good_movie, axis ="columns" ) # give it an access parameter to have it call on rows i,ie horiziontially, think of it as we are moving across the columns horizontially appossed to traditionally we traverse vertically across the rows
```
